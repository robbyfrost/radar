```python
import matplotlib.pyplot as plt
from matplotlib import rc
from matplotlib.ticker import MultipleLocator
import pyart

# plotting set up
plt.rcParams['axes.labelweight'] = 'normal'
plt.rcParams['text.latex.preamble'] = r'\usepackage{bm}'
rc('font', family='sans-serif')
rc('font', weight='normal', size=15)
rc('figure', facecolor='white')
```

```python
drad = '/data/arrcwx/robbyfrost/raxpol/20250515/'
dfile = 'RAXPOL-20250515-191108-E2.0.nc'

radar = pyart.io.read(drad + dfile)
```

```python
fig, ax = plt.subplots(1, 1, figsize=(8,5))

ax.plot(radar.time['data'], radar.elevation['data'], lw=2, label='Elevation', c="blue")
ax.set_xlabel('Time [s]')
ax.set_ylabel('Elevation [$^{\\circ}$]', c="blue")
ax.set_xlim(radar.time['data'][0], radar.time['data'][-1])
ax.grid(alpha=0.5)
ax.yaxis.set_major_locator(MultipleLocator(5))
ax.tick_params(axis='y', colors='blue')

ax2 = ax.twinx()
ax2.plot(radar.time['data'], radar.azimuth['data'], 'r', lw=2, label='Azimuth')
ax2.set_ylabel('Azimuth [$^{\\circ}$]', c="red")
ax2.yaxis.set_major_locator(MultipleLocator(90))
ax2.tick_params(axis='y', colors='red')
ax2.axvline(x=3, c='gray', ls="--")
```
